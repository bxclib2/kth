```python
def get_p_r(scores, ks, total_relevant):
    precisions = []
    recalls = []
    for cutoff in ks:
        scores_at_k = scores[:cutoff]
        relevant = scores_at_k[scores_at_k > 0].size
        precision = relevant / cutoff
        #if len(precisions):
            #precision = max(precision, np.max(precisions))
        precisions.append(precision)
        recalls.append(relevant / total_relevant)
    return precisions, recalls
```

```python
def plot_p_r(queries):
    fig, axarr = plt.subplots(1, len(queries), figsize=(15,5))
    plt.tight_layout()
    subplot = 0
    sns.set_palette(sns.color_palette("Set2", 10))
    for query in queries:
        name, scores = query
        ax = axarr[subplot]
        ax.set_xlim(-0.1, 1.1)
        ax.set_ylim(-0.1, 1.1)
        subplot += 1
        #ax.plot(recalls, precisions)
        #ax.scatter(recalls, precisions, marker='*')
        ax.set_xlabel('recall')
        ax.set_ylabel('precision')
        ax.set_title('precision at K')
        color = 'C%s' % subplot
        precisions, recalls = get_p_r(scores, ks=[3,10], total_relevant=20.)
        ax.plot(recalls, precisions, drawstyle='steps-post', label=name, color=color)
        ax.scatter(recalls, precisions, color=color)
        ax.legend()
```

```python
queries = [('zombie attack (unigram)', unigram_zombie), ('money transfer (unigram)', unigram_money), ('screen repair (unigram)', unigram_screen)]
plot_p_r(queries)
```
